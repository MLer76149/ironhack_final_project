```python
def clean_text(country_list):
    
    for country in country_list:
        filename = country+"videos_processed.csv"
        df = pd.read_csv("youtube_processed_data/"+filename)
        df["cleaned_text"] = ""
        for i, text in enumerate(df["raw_text"]):
            try:
                df.loc[i,"cleaned_text"] = clean_up(text)
            except:
                df.loc[i,"cleaned_text"] = np.nan
        df.to_csv("youtube_processed_data/"+country+"videos_cleaned.csv", index = False)
        
```

```python
def clean_up(s):

    s = s.lower()
    s = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', ' ' , s)
    s = re.sub(r'([a-z0-9+._-]+@[a-z0-9+._-]+\.[a-z0-9+_-]+)'," ", s)
    s = BeautifulSoup(s, 'lxml').get_text().strip()
    s = re.sub(r'\W+',' ',s) 
    s = re.sub(r'\d+',' ',s) 
    s = re.sub(r'\s+',' ',s)

    return s
```
