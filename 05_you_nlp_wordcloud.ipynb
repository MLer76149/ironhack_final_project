```python
def text_wordcloud(d):
    text = ""
    for key in d:
        text += (key + ", ") * d[key]
    
    return text
    
```

```python
list_remove = ["nfollow", "nsubscribe", "nfacebook", "ntwitter", "ninstagram", "nthe", "nwatch", "nlike", "nmusic", "nget", "ndie", "nmein", "nhier", "nmerci", "nabonnez", "nfor", "npour", "ncheck" ]

list_drop = ["nand", "ft", "gibt", "th"]

def clean_dict(dict_dict):
    for item in list_remove:
        if item in dict_dict:
            if item[1::] in dict_dict:
                dict_dict[item[1::]] += dict_dict[item]
            else:
                dict_dict[item[1::]] = dict_dict[item]
            dict_dict.pop(item)
    for item in list_drop:
        if item in dict_dict:
            dict_dict.pop(item)
    return dict_dict
```

```python
def final_words_cleaned(dict_dict, lang):
    df = pd.DataFrame.from_dict(dict_dict, orient="index")
    df = df.T
    word = str(lang)
    df.to_csv("final_words/finalwords_clean_"+lang+".csv")
    
```
